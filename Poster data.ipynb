```python
def create_heatmap(data, title, ax):
    correlation_matrix = data.corr()
    correlation = correlation_matrix.iloc[0, 1]
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 12}, ax=ax)
    ax.set_title(f'{title}\nCorrelation: {correlation:.2f}')

# Create a figure and axis objects for subplots
fig, axs = plt.subplots(1, 5, figsize=(20, 6))

# Create heatmaps for each continent
create_heatmap(asia_data, 'Asia', axs[0])
create_heatmap(africa_data, 'Africa', axs[1])
create_heatmap(europe_data, 'Europe', axs[2])
create_heatmap(americas_data, 'Americas', axs[3])
create_heatmap(oceania_data, 'Oceania', axs[4])

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

Asia (Correlation: 0.44): This indicates a moderate positive correlation between BMI and Depressive Disorders in Asia. As BMI increases, the prevalence of depressive disorders tends to increase as well, but the relationship is not extremely strong.

Africa (Correlation: 0.59): A correlation of 0.59 suggests a relatively strong positive correlation between BMI and Depressive Disorders in Africa. As BMI increases, there tends to be a higher prevalence of depressive disorders, and vice versa.

Americas (Correlation: 0.83): An extremely high correlation coefficient of 0.83 indicates a very strong positive correlation between BMI and Depressive Disorders in the Americas. Changes in BMI are highly associated with changes in the prevalence of depressive disorders in this region.

Oceania (Correlation: 0.77): Similar to the Americas, Oceania shows a very strong positive correlation (0.77) between BMI and Depressive Disorders. Changes in BMI are highly predictive of changes in the prevalence of depressive disorders in Oceania
